```python
def long_division(f,g):
    x = f.parent().gen()
    result = 0
    remainder = f
    quotient = 0
    while remainder.degree() >= g.degree():
        #print remainder.lt()
        #print g.lt()
        #print remainder.lt()/g.lt()
        t = remainder.lc()/g.lc() * x^(remainder.degree() -g.degree())
        quotient = quotient + t
        remainder = remainder - t*g
        
    if quotient * g + remainder == f:
        
        return quotient, remainder
    else:
        raise Exception("wrong answer...")
```
