```python
N=3
L=2

Rxw = PolynomialRing(QQ, ["w"+ str(i) for i in range(N)] + ["x" + str(m) for m in range(L)])
Fxw = Frac(Rxw)
Rq.<q> = PowerSeriesRing(Fxw)
w = Rxw.gens()[:N]
x = Rxw.gens()[N:]

if True:
    def X(g):
        return prod((g + wi for wi in w)) / prod((1-g+wi for wi in w )) / prod((1-xm*g for xm in x))
    
    h = dict()
    g = dict()
    for i in range(N):
        #optimization potential here
        h[i] = X(q.add_bigoh(3)-w[i]).reverse()
        g[i] = h[i]-w[i]
```

```python
RG.<G> = PolynomialRing(Rq)
#p=(G+w1)*(G+w2) - q*(1-G+w1)*(1-G+w2)*(1-x1*G)*(1-x2*G)

P = prod((G+wi for wi in w)) - q*prod((1-G+wi for wi in w))*prod((1-xm*G for xm in x))
P
```

```python
order = 5

E = dict()
f = dict()
S = SymmetricFunctions(QQ).e()
for i in range(0,N):
    E[i] = S[N-i].expand(N, alphabet = w) + O(q)
    
E[N] = 1

f[L] = P[N+L]

for m in range(L):
    f[m] =  O(q)
    
cur_order = 1
while cur_order < order:
    for M in range(N+L-1, N-1, -1):
        max_k = min(N, L+N-M)
        f[M-N] = -sum((E[N-k]*f[M-N+k] for k in range(1, max_k+1))) + P[M]
        
    for i in range(0,N):
        max_k = min(L, i)
        E[i] = (-sum((E[i-k]*f[k] for k in range(1, max_k+1))) + P[i]) / f[0]
    
    cur_order = cur_order + 1
```
