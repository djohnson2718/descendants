This is for the case with $N=2$ and $\ell = 2$.

```python
Rxw.<x1,x2,w1,w2> = PolynomialRing(QQ)
Fxw = Frac(Rxw)
Rq.<q> = PowerSeriesRing(Fxw)

def X(g):
    return (g+w1)*(g+w2)/(1-g+w1)/(1-g+w2)/(1-x1*g)/(1-x2*g)

h1 = X(q.add_bigoh(3)-w1).reverse()
h2 = X(q.add_bigoh(3)-w2).reverse()
g1 = h1 - w1
g2 = h2 - w2
```

This verfifies that the $g_i$ are the roots of the polynomial $p$ that we expected:

```python
RG.<G> = PolynomialRing(Rq)
p=(G+w1)*(G+w2) - q*(1-G+w1)*(1-G+w2)*(1-x1*G)*(1-x2*G)
print p
print
print p.subs({G:g1})
print p.subs({G:g2})
```

It also seems to be true, that the coefficients of 
$$\frac{p(G)}{(G-g1)(G-g2)}$$
do not have denominators.

It seems `sage` does not like to divide polynomials with exotic fields, so we reimplement it.

```python
def long_division(f,g):
    x = f.parent().gen()
    result = 0
    remainder = f
    quotient = 0
    while remainder.degree() >= g.degree():
        t = remainder.lc()/g.lc() * x^(remainder.degree() -g.degree())
        quotient = quotient + t
        remainder = remainder - t*g
        
    if quotient * g + remainder == f:
        
        return quotient, remainder
    else:
        raise Exception("wrong answer...")
```

```python
quotient, remainder = long_division(p,(G-g1)*(G-g2))

print quotient
print
print remainder
```
